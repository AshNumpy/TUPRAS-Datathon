```python
import pandas as pd 
import numpy as np 
import datetime as dt
```

```python
X = pd.read_csv('../Datasets/tupras_X_train.csv')
y = pd.read_csv('../Datasets/tupras_y_train.csv')

train = pd.merge(X, y, on='period', how='left')
train['period'] = pd.to_datetime(train['period'])
train['week'] = train['period'].dt.isocalendar().week
train['year'] = train['period'].dt.year

train.head()
```

```python
covid_cases = pd.read_csv('../Datasets/External Datasets/eng_vaka.csv')
covid_deaths = pd.read_csv('../Datasets/External Datasets/eng_deaths.csv')


covids = pd.merge(covid_cases, covid_deaths, on='tarih', how='left')
covids = covids.dropna()

covids['tarih'] = pd.to_datetime(covids['tarih'])

covids.head()
```

```python
test = pd.read_csv('../Datasets/tupras_X_test.csv')
test['period'] = pd.to_datetime(test['period'])
test.head()
```

```python
covids['week'] = covids['tarih'].dt.isocalendar().week
covids['year'] = covids['tarih'].dt.year
covids.head()
```

```python
covids.sort_values('tarih', inplace=True);covids.head()
```

```python
covids_weekly = covids.groupby(['week','year']).agg({
    'tarih':'max',
    'gunluk_vaka':'sum',
    'toplam_vaka':'sum',
    'gunluk_olum':'sum',
    'toplam_olum':'sum'
}).reset_index()
covids_weekly.sort_values('tarih', inplace=True)
covids_weekly.astype({'gunluk_vaka':'int64',
                      'toplam_vaka':'int64',
                      'gunluk_olum':'int64',
                      'toplam_olum':'int64',
                      'tarih':'datetime64[ns]'})
covids_weekly.head()
```

```python
df = pd.merge(train, covids_weekly, on=['week','year'], how='left')

df.sort_values('period', inplace=True)

df.drop(['tarih', 'week', 'year'], axis=1, inplace=True)

df = df[['period','target', 'gunluk_vaka', 'toplam_vaka', 'gunluk_olum',
       'toplam_olum', 'gasoline_imports', 'gasoline_exports', 'crude_imports',
       'crude_exports', 'gasoline_stocks', 'crude_stocks',
       'distillate_fuel_stocks', 'jet_stocks', 'propane_stocks',
       'gasoline_net_input', 'gasoline_output', 'distillate_fuel_output',
       'jet_output', 'propane_output', 'residual_fuel_output',
       'conventional_gasoline_spot_price', 'crude_brent_spot_price',
       'gasoline_future_price1', 'gasoline_future_price2',
       'gasoline_future_price3', 'gasoline_future_price4', 'gasoline_demand']]

df[['gunluk_vaka', 'toplam_vaka', 'gunluk_olum', 'toplam_olum']] = df[['gunluk_vaka', 'toplam_vaka', 'gunluk_olum', 'toplam_olum']].fillna(0)
df.tail()
```

```python
test['week'] = test['period'].dt.isocalendar().week
test['year'] = test['period'].dt.year
```

```python
df_test = pd.merge(test, covids_weekly, on=['week','year'], how='left')

df_test.sort_values('period', inplace=True)

df_test.drop(['tarih', 'week', 'year'], axis=1, inplace=True)

df_test = df_test[['period', 'gunluk_vaka', 'toplam_vaka', 'gunluk_olum',
       'toplam_olum', 'gasoline_imports', 'gasoline_exports', 'crude_imports',
       'crude_exports', 'gasoline_stocks', 'crude_stocks',
       'distillate_fuel_stocks', 'jet_stocks', 'propane_stocks',
       'gasoline_net_input', 'gasoline_output', 'distillate_fuel_output',
       'jet_output', 'propane_output', 'residual_fuel_output',
       'conventional_gasoline_spot_price', 'crude_brent_spot_price',
       'gasoline_future_price1', 'gasoline_future_price2',
       'gasoline_future_price3', 'gasoline_future_price4', 'gasoline_demand']]

df_test.head()
```

```python
df.to_csv('../Datasets/Manipulated Data/train.csv', index=False)
df_test.to_csv('../Datasets/Manipulated Data/test.csv', index=False)
```
